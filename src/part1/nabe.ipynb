かつて一世を風靡した芸人に「世界のナベアツ」という人物がいる。
それを踏まえて次の問に答えよ。(思いつきのネタ問題です)

## A: Nabeatsu of Japan

「日本のナベアツ」は自然数Nを受け取ります。次の条件のいずれかを満たすとき、アホになります。

条件1: Nが3の倍数である。
条件2: Nはある桁に3を含む。

条件を満たしているかどうか判定し、次のフォーマットを出力してください。  
アホであるとき: ((●˚⺣˚)<{受け取った数字}!!  
それ以外のとき: (・∀・)<{受け取った数字}"

出力例

入力  
1  
出力  
(・∀・)<1"

入力  
3  
出力  
((●˚⺣˚)<3!!

## B: Nabeatsu of the world
「世界のナベアツ」は自然数Nを受け取ると、1からNまでの自然数について順にアホになるかどうかを判定して出力する。
ただし、アホになる条件と、出力内容は前問と同じである。これを満たすものを出力しなさい。

出力例

入力  
3  
出力  
(・∀・)<1"  
(・∀・)<2"  
((●˚⺣˚)<3!!  

## C: Generalized Nabeatsu
「一般化ナベアツ」は自然数Nを受け取ると、1 $\leq$ X $\leq$ Nを満たすすべての自然数Xについて昇順にアホになるかどうかを判定して出力する。
アホになる条件は次の2条件のいずれかを満たすことである。
条件1: XはAの倍数である。
条件2: Xはある桁にBを含む。
N,A,Bを受け取ってN個の判定結果を標準出力する**関数**を作りなさい。

## D: Functional Nabeatsu of the Gaia

「地球のナベアツ」は自然数Nを受け取ると、1 $\leq$ X $\leq$ Nを満たすすべての自然数Xについて昇順にアホになるかどうかを判定して出力する。アホになる条件は次の通りである。  
条件: $f_1 \dots f_n$(すべてのfについて、f: $\mathbb{N}$ -> boolian)を受け取り、関数の返り値うち少なくとも1つがTrueである。  
Nと1つ以上の関数$f_1 \dots f_n$を受け取り、N個の判定結果を標準出力する関数を作りなさい。
また、自然数A,Bを受け取り、「自然数Nを受け取って前問の2つの条件についてどちらかを満たしているときにTrueを返す関数」を返す高階関数を作りなさい。
こうして作った関数をナベアツに与えて前問と同じ振る舞いをするか確認せよ。

注: Pythonでは関数をオブジェクトとして引数にとることができる(第1級関数)。また、関数を受け取るまたは関数を返す関数のことを高階関数という。高階関数は主に関数型プログラミングでは非常によく用いられる。
